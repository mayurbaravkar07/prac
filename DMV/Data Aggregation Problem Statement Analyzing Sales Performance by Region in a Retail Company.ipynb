```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df= pd.read_csv("customer_shopping_data.csv")
df
```

### Task 4 : Group data by region and calculate the total sales amount for each region.

```python
# To check the count of records grouped by region/branch of the mall

df.groupby("shopping_mall").count()
```

```python
# To check the count of records grouped by the product categories

df.groupby("category").count()
```

```python
# total sales for each mall branch
branch_sales = df.groupby("shopping_mall").sum()
```

```python
branch_sales
```

```python
combined_branch_category_sales = df.groupby(["shopping_mall", "category"]).sum()
```

```python
combined_branch_category_sales
```

```python
gender_counts = df['gender'].value_counts()

# Create a pie chart for gender distribution
plt.pie(gender_counts, labels=gender_counts.index)
plt.title("Gender Distribution")
plt.show()
```

```python
# Count the occurrences of each payment method
payment_counts = df['payment_method'].value_counts()

# Create a pie chart for payment method distribution
plt.pie(payment_counts, labels=payment_counts.index)
plt.title("Payment Method Distribution")
plt.show()
```

```python
# Count the occurrences of each category
category_counts = df['category'].value_counts()

# Create a pie chart for category distribution
plt.pie(category_counts, labels=category_counts.index)
plt.title("Category Distribution")
plt.show()
```
