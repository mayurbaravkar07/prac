### Task 6 : Aggregate the data to calculate summary statistics or derived metrics such as average sale prices by neighborhood or property type.

```python
average_prices_by_neighborhood = data.groupby('Location')['Price_Lakh'].mean()
average_prices_by_neighborhood
```

```python
average_prices_by_property_type = data.groupby('Region')['Price_Lakh'].mean()
average_prices_by_property_type
```

```python
aggregations = {
    'Price_Lakh': ['mean', 'median', 'max', 'min']
}
summary_statistics = data.groupby('Location').agg(aggregations)
```

### Task 7 : Identify and handle outliers or extreme values in the data that may affect the analysis or modeling process.


```python
Q1 = df['Price_Lakh'].quantile(0.25)
Q3 = df['Price_Lakh'].quantile(0.75)
IQR =  Q3 - Q1 


lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR


outliers = df[(df['Price_Lakh'] < lower_bound) & (df['Price_Lakh'] > upper_bound)]
```

```python
outliers
```

```python
#### If we want to remove outliers we can use the following approach 

filterd_data = df[(df['Price_Lakh'] >= lower_bound) & (df['Price_Lakh'] <= upper_bound)]
```
